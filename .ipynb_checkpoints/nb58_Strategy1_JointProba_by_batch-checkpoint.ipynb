# Write a convert&fill function and test

```python
def find_similar_entries2(entry, proba_table):

    df = pd.DataFrame()
    df['original'] = proba_table.index
    df['difference'] = df['original'].apply(np.array) - np.array(entry)
    df['non_identity'] = df['difference'].apply(np.count_nonzero)
    min_difference = np.min(df['non_identity'])
    similar_entries = df['original'][df['non_identity'] == min_difference].drop_duplicates()
    
    probas = proba_table.loc[list(similar_entries)]
    
    return np.mean(probas)

def convertNfill_proba(train_set, test_set):
    '''
    suffix 1 represent train and 2 represent test.
    '''
    tuple1 = train_set.apply(tuple, axis=1)
    tuple2 = test_set.apply(tuple, axis=1)
    tuple1.name = 'tuple'
    tuple2.name = 'tuple'
    
    jp = psjp.JointProbability()
    proba1 = jp.fit_transform(train_set, target, label='proba')
    proba2 = jp.transform(test_set, label='proba')
    
    df1 = pd.concat([tuple1, proba1], axis=1).set_index('tuple')
    df2 = pd.DataFrame(proba2)
    df2['tuple'] = tuple2
    df2.loc[df2['proba'].isnull(), 'proba'] = \
                        df2.loc[df2['proba'].isnull()]['tuple'].apply(find_similar_entries2, args=(df1,))
    
    output1 = df1['proba'].reset_index(drop=True)
    output2 = df2['proba'].reset_index(drop=True)
    
    return {'train': output1, 'test': output2}
```
