# Coming back to basics
1. Check for the independence within each category of features (e.g. binary)
2. Calculate conditional probabilities of being in class1 (or class0) given a specific value
   of each feature (continuous features to be binned prior to calculations)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
from matplotlib import cm
import statsmodels.api as sm
import seaborn as sns
import feature_processing as fp

%matplotlib inline
```

```python
train = pd.read_csv('train.csv', header=0)

binary_fs = [col for col in train.columns if '_bin' in col]
cat_fs = [col for col in train.columns if '_cat' in col]
other_fs = [col for col in train.iloc[:, 2:].columns 
              if col not in cat_fs
              if col not in binary_fs]

binaries = train[binary_fs]
categoricals = train[cat_fs]
others = train[other_fs]
```

```python
significance_level = 0.05
num_binary_fs = len(binary_fs)
association_table1 = np.zeros((num_binary_fs, num_binary_fs))

for i in range(num_binary_fs-1):
    for j in range(i+1, num_binary_fs):
        table = sm.stats.Table.from_data(binaries.iloc[:, [i, j]])
        rslt = table.test_nominal_association()
        pval = rslt.pvalue
        #if pval < significance_level:
        #    text = '  NOT  independent'
        #else:
        #    text = 'Likely independent'
        #print('(%2d, %2d) %s (p-value=%.3f)' % (i, j, text, pval))
        association_table1[i, j] = pval
        association_table1[j, i] = pval
association_table1 = pd.DataFrame(association_table1, columns=binary_fs, index=binary_fs)
```

```python
plt.figure(figsize=(12,9))
sns.heatmap(association_table1,linewidths=0.2,vmax=1.0, square=True, 
            cmap=cm.afmhot, linecolor='white', annot=True)
print('Heatmap for the independence/association test among binary features.')
print('The higher the p-value, the higher the independence.')
```

All the 'ind' variables are NOT independent of each other. 'calc' variables are mostly independent of each other except calc_15-calc_20 pair and maybe calc_16-calc17 pair. Needs to combine the associated variables into less number of more independent features.

```python
significance_level = 0.05
num_categorical_fs = len(cat_fs)
association_table2 = np.zeros((num_categorical_fs, num_categorical_fs))
data = categoricals.replace({-1:np.NaN}).dropna()

for i in range(num_categorical_fs-1):
    for j in range(i+1, num_categorical_fs):
        table = sm.stats.Table.from_data(data.iloc[:, [i, j]])
        rslt = table.test_nominal_association()
        pval = rslt.pvalue
        #if pval < significance_level:
        #    text = '  NOT  independent'
        #else:
        #    text = 'Likely independent'
        #print('(%2d, %2d) %s (p-value=%.3f)' % (i, j, text, pval))
        #print('=== (%2d, %2d) ===' % (i, j))
        #print(table.table_orig)
        association_table2[i, j] = pval
        association_table2[j, i] = pval
association_table2 = pd.DataFrame(association_table2, columns=cat_fs, index=cat_fs)
```

```python
plt.figure(figsize=(12,9))
sns.heatmap(association_table2,linewidths=0.2,vmax=1.0, square=True, 
            cmap=cm.afmhot, linecolor='white', annot=True)
print('Heatmap for the independence/association test among binary features.')
print('The higher the p-value, the higher the independence.')
```

Most features seem to be associated with each other except car_04-car05 and car_07-car_08 pairs

```python
associated_categorical_fs = ['ps_ind_02_cat', 'ps_ind_04_cat', 'ps_ind_05_cat',
                             'ps_car_01_cat', 'ps_car_02_cat', 'ps_car_03_cat',
                             'ps_car_06_cat', 'ps_car_09_cat', 'ps_car_10_cat',
                             'ps_car_11_cat']
```

```python
assoc_categoricals = []
for f in associated_categorical_fs:
    encoded = fp.encode_my_categorical_labels(categoricals[f])
    assoc_categoricals.append(encoded)
assoc_categoricals = pd.concat(assoc_categoricals, axis=1)
```

```python
assoc_categoricals
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
X = np.array(assoc_categoricals)
Xpca = pca.fit_transform(X)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
ax1.scatter(Xpca[:, 0], Xpca[:, 1], s=5, alpha=0.3)
ax2.scatter(Xpca[:, 1], Xpca[:, 2], s=5, alpha=0.3)
ax3.scatter(Xpca[:, 2], Xpca[:, 3], s=5, alpha=0.3)
```

```python
cumu_variance_ratio = []
cumu_ratio = 0
num = len(pca.explained_variance_ratio_) 
for i in range(num):
    cumu_ratio += pca.explained_variance_ratio_[i]
    cumu_variance_ratio.append(cumu_ratio)

num_to_plot=100
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
ax1.bar(range(num_to_plot), pca.explained_variance_ratio_[:num_to_plot])
ax1.grid(ls='--', lw=0.5)
ax2.axhline(y=0.95, ls='--', lw=0.5, color='0.5')
ax2.axhline(y=0.9, ls='--', lw=0.5, color='0.5')
ax2.axhline(y=0.75, ls='--', lw=0.5, color='0.5')
ax2.axhline(y=0.5, ls='--', lw=0.5, color='0.5')
ax2.bar(range(num_to_plot), cumu_variance_ratio[:num_to_plot])
```
