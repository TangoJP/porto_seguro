```python
train = pd.read_csv('train.csv', header=0)
test = pd.read_csv('test.csv', header=0)

all_fs = train.columns[2:]

binary_fs = sorted([f for f in all_fs if '_bin' in f])
categorical_fs = sorted([f for f in all_fs if '_cat' in f])
other_fs = sorted([f for f in all_fs
                if f not in binary_fs
                if f not in categorical_fs])

target = train.target
```

```python
used_fs = other_fs.copy()
used_fs.remove('ps_calc_01')
used_fs.remove('ps_calc_02')
used_fs.remove('ps_calc_03')
used_fs.remove('ps_calc_04')
used_fs.remove('ps_calc_06')
used_fs.remove('ps_calc_08')
used_fs.remove('ps_calc_09')
print(used_fs)
```

```python
data_train = train[used_fs]
data_test = test[used_fs]
```

```python
train_all = data_train[used_fs]
train_notnull = train_all.replace({-1:np.NaN}).dropna()
train_null_index = [i for i in train_all.index if i not in train_notnull.index]
train_null = train_all.iloc[train_null_index]
```

```python
test_all = data_test[used_fs]
test_notnull = test_all.replace({-1:np.NaN}).dropna()
test_null_index = [i for i in test_all.index if i not in test_notnull.index]
test_null = test_all.iloc[test_null_index]
```

```python
print('==== Train Set Missing Values ====')
print('# All Entries: %d' % len(train_all))
print('# Entries w/o missing data: %d' % len(train_notnull))
print('# Entries w/ missing data: %d' % (len(train_null)))
print('\n=== # Missing Values For Each Fature')
for i, f in enumerate(train_all.columns):
    feature = train_all[f]
    print('%s: %d' % (f, len(feature[feature == -1])))
```

```python
print('==== Test Set Missing Values ====')
print('# All Entries: %d' % len(test_all))
print('# Entries w/o missing data: %d' % len(test_notnull))
print('# Entries w/ missing data: %d' % (len(test_null)))
print('\n=== # Missing Values For Each Fature')
for i, f in enumerate(test_all.columns):
    feature = test_all[f]
    print('%s: %d' % (f, len(feature[feature == -1])))
```

```python
pca = PCA()
Xpca_train = pca.fit_transform(np.array(train2))
Xpca_test = pca.transform(test2)

pca_labels = [('component_' + str(i)) for i in range(Xpca_train.shape[1])]
train_pca = pd.DataFrame(Xpca_train, columns=pca_labels)
test_pca = pd.DataFrame(Xpca_test, columns=pca_labels)
```

```python
pca_pair_list = list(combinations(train_pca.columns, 2))
pca_pair_list1 = pca_pair_list[:80]
pca_pair_list2 = pca_pair_list[80:]

print(len(pca_pair_list1))
print(len(pca_pair_list2))
```

```python
num_plot= 20000
fig = plt.figure(figsize=(18, 42))
for i, pair in enumerate(pca_pair_list1):
    x0 = np.array(train_pca[pair[0]])
    x1 = np.array(train_pca[pair[1]])
    
    ax = fig.add_subplot(14, 6, i+1)
    ax.scatter(x0[:num_plot], x1[:num_plot], color='yellowgreen', alpha=0.1)
    ax.set_xlabel(pair[0])
    ax.set_ylabel(pair[1])
plt.tight_layout()
```
