```python
train = pd.read_csv('train.csv', header=0)

all_fs = train.columns[2:]
binary_fs = sorted([f for f in all_fs if '_bin' in f])
categorical_fs = sorted([f for f in all_fs if '_cat' in f])
other_fs = sorted([f for f in all_fs
                if f not in binary_fs
                if f not in categorical_fs])

target = train.target
```

```python
num_samples = len(target)
num_target = np.sum(target)
freq_target = num_target/num_samples
freq_target
```

# Binaries

### ind_bin

```python
ind_bins_fs = [f for f in all_fs 
               if '_bin' in f
               if '_ind' in f]
ind_bins = train[ind_bins_fs]

ind_bins.columns
```

```python
total = 0
for i in range(1, len(ind_bins_fs)+1):
    total += sc.misc.comb(len(ind_bins_fs), i)
print(int(total))
```

```python
ind_bins_fs2 = ind_bins_fs.copy()
ind_bins_fs2.remove('ps_ind_08_bin')
ind_bins_fs2.remove('ps_ind_09_bin')
ind_bins2 = train[ind_bins_fs2]
```

```python
total = 0
for i in range(1, len(ind_bins_fs)+1):
    total += sc.misc.comb(len(ind_bins_fs2), i)
print(int(total))
```

```python
%%time
optimizer1 = psjp.JointProbabilityOptimizer(ind_bins2, verbose=True)
res1 = optimizer1.exhaustiveCombinatorialSelection(target, parallel=True)
```

```python
res1
```

```python
print(optimizer1.best_expected_value_)
print(optimizer1.best_combination_)
```

After all combining all of them yields the best result
