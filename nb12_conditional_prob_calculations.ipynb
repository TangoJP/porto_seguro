```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
import seaborn as sns

from matplotlib import cm
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
%matplotlib inline
```

```python
data = pd.read_csv('./data/trainset_digitized.csv')
```

```python
data.head(10)
```

```python
target = data.target
class0_proba_data = pd.DataFrame()
class1_proba_data = pd.DataFrame()
class1_odss_data = pd.DataFrame()

for i, feature in enumerate(data.columns[:-1]):
    vec = data.iloc[:, i]
    feature_space = vec.unique()
    probas_class0 = {}
    probas_class1 = {}
    odds_of_class1 = {}
    for j, val in enumerate(feature_space):
        data_val = len(vec[vec == val])
        data_val_class0 = len(vec[(vec == val) & (target == 0)])
        data_val_class1 = len(vec[(vec == val) & (target == 1)])
        if data_val == 0:
            prob_class0_given_val = 0
            prob_class1_given_val = 0
            odds = 0
        else:
            prob_class0_given_val = data_val_class0 / data_val
            prob_class1_given_val = data_val_class1 / data_val
            if prob_class0_given_val == 0:
                odds = 0
            else:
                odds = prob_class1_given_val / prob_class0_given_val

        probas_class0[val] = prob_class0_given_val
        probas_class1[val] = prob_class1_given_val
        odds_of_class1[val] = odds
    
    class0_proba_data[feature] = vec.replace(probas_class0)
    class1_proba_data[feature] = vec.replace(probas_class1)
    class1_odss_data[feature] = vec.replace(odds_of_class1)
```
