```python
total = len(train)
target_0 = len(train[train.target == 0])
target_1 = len(train[train.target == 1])
print('%% of Target 0: %6.3f%%' % (100*target_0/total))
print('%% of Target 1: %6.3f%%' % (100*target_1/total))
```

```python
train_features = train.iloc[:, 2:]
```

```python
binary_fs = [col for col in train.columns if '_bin' in col]
binary_fs
```

```python
non_cat_fs = [col for col in train_features.columns 
              if col not in cat_fs
              if col not in binary_fs]
```

```python
non_cat_fs
```

```python
X = np.array(train[non_cat_fs])
scaler = Normalizer()
X = scaler.fit_transform(X)

pca = PCA()
Xpca = pca.fit_transform(X)
```

```python
Xpca.shape
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(Xpca[:, 0], Xpca[:, 1], s=5, alpha=0.5, c=train.target, cmap=cm.Reds)
```

```python
vs = pca.explained_variance_ratio_
vs_cumu = []
for i in range(len(vs)):
    if i == 0:
        vs_cumu.append(vs[i])
    else:
        vs_cumu.append(np.sum(vs[:i]))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
ax1.bar(range(len(vs)), vs)
ax2.bar(range(len(vs)), vs_cumu)
ax2.axhline(y=0.25, c='yellowgreen', ls='--', lw=1)
ax2.axhline(y=0.5, c='yellowgreen', ls='--', lw=1)
ax2.axhline(y=0.75, c='yellowgreen', ls='--', lw=1)
ax2.axhline(y=0.9, c='yellowgreen', ls='--', lw=1)
ax2.axhline(y=0.95, c='yellowgreen', ls='--', lw=1)
ax1.set_xlabel('Component')
ax1.set_ylabel('Explained Variance')
ax2.set_xlabel('Component')
ax2.set_ylabel('Cumulative Explained Variance')
```
